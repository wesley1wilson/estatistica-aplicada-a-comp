```python
import pandas as pd
import matplotlib.pyplot as plt
titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

Primeiramente, eu adicionei as bibliotecas para que a gente pudesse abrir o arquivo requisitado e em seguida o matplotlib para poder gerar os gráficos.

```python
a1 = titanic['Survived'].value_counts(normalize=True)*100
a2 = titanic['Pclass'].value_counts(normalize=True)*100
a3 = titanic['Sex'].value_counts(normalize=True)*100
a4 = titanic['Embarked'].value_counts(normalize=True)*100
```

Em seguida, eu transformei cada variável que se classificasse em variável qualitativa e que fosse possível gerar o gráfico e converti para porcentagem.

Logo após, como pode ver em seguida, para reproduzir o gráfico eu usei o comando plot mais um complemento (pie ou bar) e indiquei o tamanho e o formato dos números. No caso do gráfico de rosca foi preciso criar um outro círculo em branco para dar o aspecto de buraco no meio, E depois só adicionei o título.

<b>GRÁFICO EM SETORES</b>

```python
a1.plot.pie(figsize=(6, 6), autopct='%1.1f%%')
plt.title('Survived')
```

```python
a2.plot.pie(figsize=(6, 6), autopct='%1.1f%%')
plt.title('Pclass')
```

```python
a3.plot.pie(figsize=(6, 6), autopct='%1.1f%%')
plt.title('Sex')
```

```python
a4.plot.pie(figsize=(6, 6), autopct='%1.1f%%')
plt.title('Embarked')
```

<b>GRÁFICO DE ROSCA</b>

```python
circulo=plt.Circle( (0,0), 0.7, color='white')
a1.plot.pie(y='Survived', figsize=(5, 5))
p=plt.gcf()
p.gca().add_artist(circulo)
plt.title('Survived')
```

```python
circulo=plt.Circle( (0,0), 0.7, color='white')
a2.plot.pie(y='Pclass', figsize=(5, 5))
p=plt.gcf()
p.gca().add_artist(circulo)
plt.title('Pclass')
```

```python
circulo=plt.Circle( (0,0), 0.7, color='white')
a3.plot.pie(y='Sex', figsize=(5, 5))
p=plt.gcf()
p.gca().add_artist(circulo)
plt.title('Sex')
```

```python
circulo=plt.Circle( (0,0), 0.7, color='white')
a4.plot.pie(y='Embarked', figsize=(5, 5))
p=plt.gcf()
p.gca().add_artist(circulo)
plt.title('Embarked')
```

```python
a4.plot.bar()
plt.title('Embarked')
```
